```python
from konlpy.tag import Okt
okt = Okt()
def okt_tokenizer(x):
    return okt.morphs(x)
```

# Data

```python
review=pd.read_csv('review107.csv')
```

```python
review['review']=review['review'].apply(lambda x : cleanText(x))
```

```python
tf_data = review.groupby(['name','prof'])['review'].agg(lambda x : ''.join(x)).reset_index()
corpus = tf_data['review'].tolist()
```

## 1.TDM

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(tokenizer=okt_tokenizer, min_df = 0.3)
vect.fit(corpus)

TDM = vect.transform(corpus).toarray()
```

## 2. TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(tokenizer=okt_tokenizer, min_df=0.1).fit(corpus)
tfidf_matrix = tfidf.transform(corpus).toarray()

tfidf_matrix.shape
```

```python
tfidf_matrix
```

## 3. Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python
tf_data.query('prof=="유재명"')
```

```python
corpus[1448]
```

```python
tfidf_matrix[1448,:]
```

```python
A=sparse.csr_matrix(tfidf_matrix)
```

```python
B=cosine_similarity(A)
```

```python
np.argsort(B[405,:])[::-1][1]
```

```python
name = '세계역사의이해'
prof = '윤은주'

idx = tf_data.query('name == @name & prof == @prof').index[0]
sim_idx = np.argsort(B[idx,:])[::-1][1]
display(tf_data.loc[sim_idx,'name'])
display(tf_data.loc[sim_idx,'prof'])
display(tf_data.loc[sim_idx,'review'])
```
