```python
import pandas as pd
import numpy as np
```

```python
import re
 
def cleanText(readData):
    text = re.sub('[^가-힣]',' ',readData)
    return text
```

```python
from konlpy.tag import Okt
okt = Okt()
def okt_tokenizer(x):
    return okt.morphs(x)
```

# Data

```python
review=pd.read_csv('review103.csv')
```

```python
review['review']=review['review'].apply(lambda x : cleanText(x))
```

```python
tf_data = review.groupby(['name','prof'])['review'].agg(lambda x : ''.join(x)).reset_index()
corpus = tf_data['review'].tolist()
```

```python
(review[:10].groupby(['name','prof'])['review'].agg(lambda x : ' '.join(x)).reset_index())['re']
```

## 1.TDM

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(tokenizer=okt_tokenizer, min_df = 0.3)
vect.fit(corpus)

TDM = vect.transform(corpus).toarray()
```

## 2. TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(tokenizer=okt_tokenizer, min_df=0.2).fit(corpus)
tfidf_matrix = tfidf.transform(corpus).toarray()

tfidf_matrix.shape
```

```python
tfidf_matrix
```

## 3. Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python
tf_data.query('prof=="유재명"')
```

```python
corpus[405]
```

```python
tfidf_matrix[405,:]
```

```python
A=sparse.csr_matrix(tfidf_matrix)
```

```python
B=cosine_similarity(A)
```

```python
np.argsort(B[405,:])[::-1][1]
```

```python
def find_similar(in_name,in_prof):
    idx = tf_data.query('name == @in_name & prof == @in_prof').index
    sim_idx = np.argsort(B[idx,:])[::-1][1]
    tf_data[sim_idx]
```
